```python
df['Emergency_Fund'] = df['salary']*0.10
df
```

```python
def ef(num):
    if(num<10000):
        return('Not_Enough')
    else:
        return('Enough')
```

```python
df['Fund_Status'] = df['Emergency_Fund'].apply(ef)
df
```

```python
sns.countplot(x='Fund_Status',data=df,hue = 'level of education')
```

```python
#onfidence interval is a range that defines what is the probability of the correct value
import numpy as np
sns.barplot(x='division', y='salary',data =df,estimator= np.mean,errorbar = 'sd')
```
