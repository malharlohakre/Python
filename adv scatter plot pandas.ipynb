```python
df['Total_Exp']= df['TV'] + df['radio'] + df['newspaper']
```

```python
df = df.drop('Total_Exp',axis = 1) #to delete the column 
```

```python
df['Total_kharcha']= df['TV'] + df['radio'] + df['newspaper']
```

```python
df
```

```python
def cat(x):
    if (x>300):
        return('high')
    elif(x<299 and x>=150):
        return('med')
    else:
        return('low')
cat(100)
```

```python
df['Exp_cat'] = df['Total_kharcha'].apply(cat)
df
```

```python
sns.scatterplot(data = df,x='Total_kharcha',y='sales',hue = 'Exp_cat',palette = 'Dark2',size='newspaper',style = 'TV' )
#x = x-axis , y = y-axis, hue = it will add color and palette will add colur of your own choice for that search on google colour maps in matplotlib ther you will see code which you have to add
#shift tab it will show what you need in formula 
#scatter flot will show realtion it requires two axis
```
