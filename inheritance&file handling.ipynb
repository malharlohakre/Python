```python
class dr322():#parent class
    def __init__(self,n1):
        self.n1 = n1
    def add_values(self):
        sum=0
        for i in self.n1:
            sum = sum+i
        print('the total is',sum)
list1 = [1,2,3]
ob1 =dr322(list1)
ob1.add_values()


class dr322_1(dr322):#child class
    pass
ob2 = dr322_1(list1)
print(type(ob2))
print(type(ob1))
ob2.add_values()
#object of one class cannot use mwthod of another class
#object of one class can acess the methods of another class only if there is a parent child relationship between the classes
#one parent class can have the multiple child classes
#if the child has its own attributes and methods it cannot access the method of the parent class
class dr322_3(dr322):
    pass
ob3 = dr322_3(list1)
ob3.add_values()

class dr322_4():
    def __init__(self,a):
        self.a = a
    def ret_val(self):
        print(self.a)
ob10 = dr322_4(5)
ob10.ret_val()

```
