```python
class human():#class

    def __init__(self, name, age):#this is to be used when we have to define attributes
        self.name = name
        self.age = age
    def can_dance(self):#method
        print(self.name, "can dance", self.age, 'years old')
    def can_sing(self):#method
        print(self.name,'can sing', self.age, 'years old')
    def can_walk(self):#method
        print(self.name, 'can walk', self.age, 'years old')

human1 = human('malhar', 22)#objects
human2 = human('kirti', 25)#objects
human3 = human('vaibhav', 31)#objects
human4 = human('gupta', 50)#objects
print(type(human2))
#connect the method to the object
human1.can_dance()#method to object connection
human1.can_sing()
human4.can_walk()
```

```python
class math_operation():
    def __init__(self,n1,n2,str1):
        self.n1 = n1
        self.n2 = n2
        self.str1 = str1

    def add(self):
        result = self.n1+self.n2
        print("the total of",self.n1,"and",self.n2,"is",result)

    def sub(self):
        result = self.n1-self.n2
        print("the subtraction of",self.n1,"and",self.n2,"is",result)

    def find_letters(self):
        c=0
        for i in self.str1:
            c+=1
        print('the no of characters are',c)

    def even_odd_first_val(self):
        if(self.n1%2==0):
            print(self.n1,"is even")
        else:
            print(self.n1,"is odd")

    def even_odd_second_val(self):
        if(self.n2%2==0):
            print(self.n2,"is even")
        else:
            print(self.n2,"is odd")

    def max_num(self):
        if (self.n1 > self.n2):
            print(self.n1,'is grater')
        elif:
            print(self.n1,'is smaller')
        else:
            print('they are equal')

ob1 = math_operation(12,3,'i am learning OOPs')
ob1.add()
ob1.sub()
ob1.find_letters()

ob2 = math_operation(44,7,"hello")
#ob2.add()
ob2.even_odd_first_val()
ob2.even_odd_second_val()
ob2.max_num()
#tuuple cant be modified 
```

```python
#rules of oops

class math_op1():
    def __init__(self,n1,n2):
        self.n1 = n1
        self.n2 = n2
    def add(self):
        print('the total is',self.n1+self.n2)
    def sub(self):
        print('the diff is',self.n1 - self.n2)

ob1 = math_op1(1,2)
ob1.add()
class math_op2(math_op1):
    pass
ob2 = math_op2(5,5)
# we have to diff class
#object of on1 class cannot access method of another class
#object of one class can access method of another class provided therr is parent child relatonship between them
ob2.add()
ob2 = math_op2(6,4)
ob2.sub()

class math_op3():
    pass
ob3 = math_op3(4,5)
ob3.sub()
```
