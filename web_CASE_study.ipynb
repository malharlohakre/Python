```python
#https://books.toscrape.com/
import requests
from bs4 import BeautifulSoup
import csv
import re
```

```python
url = 'https://books.toscrape.com/'
response = requests.get(url)

if response.status_code != 200:
    print('failed to retrive the webpage')
    exit()
```

```python
# parse the HTML content
data = BeautifulSoup(response.text,'html.parser')
#print(data)
```

```python
books = data.find_all('article', class_ = 'product_pod')
book_list = []

for book in books:
    title = book.find('h3').find('a')['title']
    price = book.find('p',class_='price_color').text
    price_matched = re.search(r'Â£([0-9]+\.[0-9]+)',price)
    price_value = float(price_matched.group(1)) if price_matched else None
    book_list.append({'Title': title, 'Price':price_value})

print(book_list)
```

```python
# save file as csv
with open('books.csv','w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Title','Price'])
    writer.writerows([[book['Title'], book['Price']] for book in book_list])
```
