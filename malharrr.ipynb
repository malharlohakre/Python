```python
#10
import random

def hangman():
    words = ['malhar','mith','chicken','lolipop','project']
    word = random.choice(words)
    guessed = set()
    tries = 7
    name = input('enter your name')
    
    while tries > 0:
        show =''
        for letter in word:
            if letter in guessed:
                show += letter
            else:
                show += 'x'
        print('word',show)
    
        guess = input('guess a letter').lower()
        
        if guess in word:
            guessed.add(guess)
            print('correct')
        else:
            tries -= 1
            print('wrong , tries left',tries)
        if show == word:
            print('you win')
            result = 'win'
            break

    else:
        print('game over the word was',word)
        result = 'lose'
hangman()
                
    
```
