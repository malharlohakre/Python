```python
#regular expression 
#a sequence of characters forming a search pattern. used for pattern matching and text manipukation


special characters
. it matches any character
^ it matches start of string
$ it mtches end of string

* more 0 or more occurancce
+ for 1 0r more occurance
? for 0 or 1 occurance

[abc] it matches a,b,or c
[^abc] it matches any except a,b,c
/d it matches any digit
```

```python
#regEx : regular expression
import re
text = "cat rat bat malhar"
pattern = ".at"

result = re.findall(pattern,text)
print(result)

text1 = "hello malhar"
text2 = "malhar"
pattern1 = "^hello"
result = re.findall(pattern1,text1)
print(result)


text1 = "wlcome malhar"
text2 = "malhar"
pattern3 = "ar$"
result = re.findall(pattern3,text1)
print(result)

text8 = "malhar is eating double cheese chicken extra pizza"
pattern3 = "pizza$"
result = re.findall(pattern3,text8)
print(result)
```
