```python
from bs4 import BeautifulSoup #to import
```

```python
f = open("C:\\sample.html","r",encoding = "utf-8") # encoding whatever written i the website python will read it
html = f.read()
f.close()
```

```python
soup = BeautifulSoup(html,"html.parser")#parser point at each and evry character 
articles = soup.find_all("div",class_ = "article")
```

```python
for article in articles:
    headline = article.find("h2").text #h2 to define tag () iske beech mai jo likhenge wo artcile hai
    author = article.find("span",class_="author").text #span -sub information
    date = article.find("span",class_="date").text
    print('headline',"-",author,"-",date)
```
