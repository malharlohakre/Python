```python
def subtract(a,b):
    if(b==0):
        return(a)
    return subtract (a-1,b-1)

stock = 10
price_per_item = 100
print('welcome to malhar shopping mart')
while stock > 0:
    print('current stock',stock)
    qty = int(input('enter the quantity'))
    if(qty>stock):
        print('not enough stock')
    else:
        stock = subtract(stock,qty)
        bill = qty*price_per_item
        print('bill for this purchase',bill)
        print('stock for this purcahse',stock)
print('out of stock')
```
